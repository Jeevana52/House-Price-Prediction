```python
from sklearn.model_selection import train_test_split
X = HouseDF.drop(['squareMeters'], axis=1)
Y = HouseDF['squareMeters']
```

```python
Y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y,test_size=0.4)
```

```python
train_HouseDF=X_train.join(Y_train)
```

```python
train_HouseDF
```

```python
plt.figure(figsize=(25,12))
sns.heatmap(train_HouseDF.corr(), annot=True, cmap="PiYG")
```

```python
train_HouseDF['numberOfRooms']=np.log(train_HouseDF['numberOfRooms']+1)
train_HouseDF['garage']=np.log(train_HouseDF['garage']+1)
train_HouseDF['hasGuestRoom']=np.log(train_HouseDF['hasGuestRoom']+1)
train_HouseDF['basement']=np.log(train_HouseDF['basement']+1)
```

```python
train_HouseDF.hist(figsize=(25,12))
```

```python
plt.figure(figsize=(25,12))
sns.scatterplot(x="garage", y="basement", data=train_HouseDF, hue="numberOfRooms", palette="rocket")
```
